```python
import csv
import numpy as np
import scipy.stats
import scipy as sp
import re
import matplotlib.pyplot as plt
import matplotlib
from tabulate import tabulate

ispdf = False
path = "../results/"
files = [
    "results-9-23-2022.csv",
    "results-9-30-2022.csv",
    "results-9-30-2022-temp.csv",
    "results-1-5-2023.csv"
]

graph_name = [
    "livej",
    "patent", 
    "wikitalk",
    "roadNetCA",
    "dbpedia",
    "baidu",
    "pokec",
    "wiki_talk_en",
    "wiki-links",
    #"ssca2",
    # "com-friendster",
    # "twitter",
    # "twitter_mpi",
    "er-1m-8m",
    "ba-1m-8m",
    "rmat-1m-8m",
    "DBLP",
    "Flickr",
    "StackOverflow",
    "wiki-edits-sh",
]

# for stability test or scalability test
#selected_graph = {"wikitalk", "roadNetCA", "dbpedia", "ba-1m-8m"}

graph_title = [
    "livej",
    "patent",
    "wikitalk",
    "roadNet-CA",
    "dbpedia",
    "baidu", 
    "pokec",
    "wiki-talk-en",
    "wiki-links-en",
    #"ssca2",
    # "com-friendster",
    # "twitter",
    # "twitter-mpi",
    "ER",
    "BA",
    "RMAT",
    "DBLP",
    "flickr",
    "StackOverflow",
    "wiki-edits-sh"
]


workers = ["1", "2", "4", "8", "16", "32", "64"]

g_x = [1, 2, 4, 8, 16, 32, 64]
data = [] # load all data in csv files

def read_data():
    global data
    data = []
    for p in files:
        p = path + p
        f = open(p,'rt')
        reader = csv.DictReader(f)
        for row in reader:
            data.append(row)
        f.close()
```

```python

def mean_confidence_interval(data, confidence=0.95):
    a = 1.0*np.array(data)
    n = len(a)
    m, se = np.mean(a), scipy.stats.sem(a)
    h = se * sp.stats.t._ppf((1+confidence)/2., n-1)
    return round(m, 2), round(h, 2), round(h, 2)

def parse_data_time(model, worker, alg, isInsert):
    global data
    localtimes = []
    for row in data:
        if model != row["model"]: continue
        if worker != row["workers"]: continue
        if alg != row["alg"]: continue
        if alg == "4": 
            if isInsert and row["insert-num"] != "0": 
                if row["mstime"] != "-1": localtimes.append(float(row["mstime"]))
            if not isInsert and row["remove-num"] != "0":
                if row["mstime"] != "-1":localtimes.append(float(row["mstime"]))
        if alg == "41" or alg == "43" or alg == "42":
            if isInsert and row["Insert"] != "-1": 
                localtimes.append(float(row["Insert"]))
            if not isInsert and row["Delete"] != "-1":
                localtimes.append(float(row["Delete"]))
    if model == "baidu" and worker == "1": 
        print(alg); print(isInsert)
        print(localtimes); print("")
    if len(localtimes) != 0 : 
        return mean_confidence_interval(localtimes)
    else: return (0, 0)

    
# draw the running time by varying the number of workers
def draw_running_time_by_worker_number():
    global graph_name; global graph_title; global g_x
    index = 1
    #fig = plt.figure(figsize=(16, 12))
    fig = plt.figure(figsize=(16, 22))
    
    for model, title in zip(graph_name, graph_title):
        # ax = fig.add_subplot(4, 4, index); index += 1
        ax = fig.add_subplot(6, 3, index); index += 1
        
        y1 = []; y2 = []; y3 = []; y4 = []; y5 = []; y6 = []; 
        z1 = []; z2 = []; z3=[]; z4 = [] #baseline for orderbased, and traversal
        y1err = []
        x = g_x
        for w in workers:
            r1 = parse_data_time(model, w, "4", True); y1.append(r1[0]); y1err.append(r1[1]) #our insert
            r2 = parse_data_time(model, w, "4", False); y2.append(r2[0]) #our remove
            r3 = parse_data_time(model, w, "41", True); y3.append(r3[0]) # other insert
            r4 = parse_data_time(model, w, "41", False); y4.append(r4[0]) # other remove
            r5 = parse_data_time(model, w, "42", True); y5.append(r5[0]) # other Insert
            r6 = parse_data_time(model, w, "42", False); y6.append(r6[0]) # other remove
        #print(x); print(y2)
        ax.plot(x, y1, label="OurI", marker="o", color='red'); #plt.yscale("log")
        #ax.errorbar(x, y1, yerr = y1err, label="OurI", marker="o")
        ax.plot(x, y2, label="OurR", marker="o", linestyle='dotted', color='red'); #plt.yscale("log")
        ax.plot(x, y3, label="JEI", marker="v", color='green'); #plt.yscale("log")
        ax.plot(x, y4, label="JER", marker="v", linestyle='dotted', color='green'); #plt.yscale("log")
        ax.plot(x, y5, label="MI", marker="^", color='blue'); #plt.yscale("log")
        ax.plot(x, y6, label="MR", marker="^", linestyle='dotted', color='blue'); #plt.yscale("log")
        
        
        #baseline order
        for w in workers: 
            z1.append(y1[0]); z2.append(y2[0])
        ax.plot(x, z1, label="OI", color = 'red', lw="1")
        ax.plot(x, z2, label="OR", color = 'red', lw="1", linestyle="dotted")
         
        
        #baseline Traversal
        r5 = parse_data_time(model, "1", "43", True)
        if r5: 
            t = r5[0]
            if t < 3000000: #only show less than 3000s data 
                for w in workers: z3.append(t)
                print(z3)
                ax.plot(x, z3, color = 'black', marker = "+", lw="1", label="TI")
                #ax.plot(x, z3, color = 'black', lw="1", linestyle="dotted")
        r6 = parse_data_time(model, "1", "43", False)
        if r6: 
            t = r6[0]
            if t < 3000000: 
                for w in workers: z4.append(t)
                print(z3)
                ax.plot(x, z4, color = 'black', marker = "+", lw="1", linestyle = 'dotted', label="TR")
                #ax.plot(x, z3, color = 'black', lw="1", linestyle="dotted")

        ax.grid()
        #ax.legend(fontsize="small", ncol=4)
        ax.set_xscale("log")
        ax.set_yscale("log")
        ax.set_title(title)
        ax.set_xticks(g_x, x) 
        
        ax.set_xlabel("Number of Workers")
        
        ax.set_ylabel("Running Time(ms)")
        
        if model=="wikitalk": handles, labels = ax.get_legend_handles_labels()
    fig.align_labels()
    fig.tight_layout()
     
    
    fig.legend(handles, labels, fontsize="18", loc="upper center", bbox_to_anchor=(0.5, 1.02), ncol=10)

    #plt.savefig("par-cm-running-time-2.pdf", bbox_inches='tight')
    plt.savefig("par-cm-running-time-3.pdf", bbox_inches='tight')
    plt.show()  

read_data(); draw_running_time_by_worker_number()
```

### draw the speedup table

```python
# draw the table of running time speedups 
# 1 worker(sequential) 16 workers running time (ms) for both methods show speedups.
# draw the running time by varying the number of workers
def draw_running_time_speedup_table():
    global graph_name; global graph_title
    lines = []
    for model, title in zip(graph_name, graph_title):
        line = [title]
        w = "1"
        r1 = parse_data_time(model, w, "4", True);  #our Insert
        r2 = parse_data_time(model, w, "4", False); #our remove
        r3 = parse_data_time(model, w, "41", True);  # JEI insert
        r4 = parse_data_time(model, w, "41", False);  # JER remove
        r5 = parse_data_time(model, w, "42", True);  # MI insert
        r6 = parse_data_time(model, w, "42", False);  # MR remove
        
        w = "16"
        s1 = parse_data_time(model, w, "4", True); #our insert
        s2 = parse_data_time(model, w, "4", False); #our remove
        s3 = parse_data_time(model, w, "41", True);  # other insert
        s4 = parse_data_time(model, w, "41", False);  # other remove
        s5 = parse_data_time(model, w, "42", True);  # other insert
        s6 = parse_data_time(model, w, "42", False);  # other remove
        
        #speedup 16 workers vs 1 worker
        speedup = str(round(r1[0]/s1[0], 1)); line.append(speedup) # ourI
        speedup = str(round(r2[0]/s2[0], 1)); line.append(speedup) # ourR
        speedup = str(round(r3[0]/s3[0], 1)); line.append(speedup) # JEI
        speedup = str(round(r4[0]/s4[0], 1)); line.append(speedup) #JER
        if s5[0] == 0: line.append(0) 
        else: speedup = str(round(r5[0]/s5[0], 1)); line.append(speedup) # MI
        if s6[0] == 0: line.append(0)
        else: speedup = str(round(r6[0]/s6[0], 1)); line.append(speedup) # MR
        
#         #speedup 1 worker our vs JE
#         speedup = str(round(r3[0]/r1[0], 1)); line.append(speedup) #ourI 
#         speedup = str(round(r4[0]/r2[0], 1)); line.append(speedup) #ourR 
        
#         #speedup 16 worker our vs JE
#         speedup = str(round(s3[0]/s1[0], 1)); line.append(speedup) #ourI 
#         speedup = str(round(s4[0]/s2[0], 1)); line.append(speedup) #ourR 
        
        #speedup 1 worker ourI vs JEI MI
        speedup = str(round(r3[0]/r1[0], 1)); line.append(speedup) 
        speedup = str(round(r5[0]/r1[0], 1)); line.append(speedup) 
        
        #speedup 1 worker ourR vs JER MR
        speedup = str(round(r4[0]/r1[0], 1)); line.append(speedup) 
        speedup = str(round(r6[0]/r1[0], 1)); line.append(speedup) 
        
        #speedup 16 worker ourI vs JEI MI
        speedup = str(round(s3[0]/s1[0], 1)); line.append(speedup) 
        speedup = str(round(s5[0]/s1[0], 1)); line.append(speedup) 
        
        #speedup 16 worker ourR vs JER MR
        speedup = str(round(s4[0]/s1[0], 1)); line.append(speedup) 
        speedup = str(round(s6[0]/s1[0], 1)); line.append(speedup) 
        
        
        lines.append(line)
        #print(line)
        
    print(tabulate(lines)) 
    
    for i in range(1, 15): #colum
        min = 100000.0; max = 0.0; min_idx = -1; max_idx = -1
        for j in range(16): #row
            a = float(lines[j][i])
            if a == 0: 
                lines[j][i] = "-"; continue
            if a < min: min = a; min_idx = j
            if a > max: max = a; max_idx = j
        lines[min_idx][i] = "\\textbf{" + str(lines[min_idx][i]) + "}"
        lines[max_idx][i] = "\\textbf{" + str(lines[max_idx][i]) + "}"
        
    for line in lines: print(" & ".join(line) + " \\\\")
        
    print("")
    print("")    
draw_running_time_speedup_table()  
```
